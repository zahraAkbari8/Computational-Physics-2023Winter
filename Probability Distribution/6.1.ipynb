```julia
using LinearAlgebra
using Plots
using Random
using Statistics
using LaTeXStrings
```

```julia
N = 100000
step = 10
CountList = zeros(Int, 10)
linelist = ones(Int, 10)
stdlist = zeros(Float64, Nstep)

xlist = collect(1:10)
for i in 1:N                            # Counting the numbers of time a value is appeared since we are not going to use histogram function for plotting
    temp = rand((1:10))
    CountList[temp] = CountList[temp] + 1
    if i%10 == 0
        stdlist[i10] = std(CountList)/i
    end
end

p1 = bar( xlist , CountList , xlabel = L"number" ,ylabel = L"count", title = L"uniform\ distribution\ histogram\ for\ N=100000" ) 
p2 = plot!(xlist, linelist*(N/10))

display(p2)
savefig("images\\6.1.png")

```

```julia

p3 = plot(1:step:N ,stdlist)        # We should notice that one set of random lists is used for plotting
p4 = plot!(x -> 1/sqrt(x), ylim =[0,0.25] , xlim = [0,2.5*10^4] , ylabel = L"STD" ,  xlabel = L"Number" , title = L"standard\ deviation\ by\ incresing\ N ")         #fitting Line
display(p3)savefig("images\\6.1prime.png")
```
