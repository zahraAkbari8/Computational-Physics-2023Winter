```julia
using LinearAlgebra
using Plots
using Random
using Statistics
using Distributions
using StatsPlots
using LaTeXStrings
```

```julia
N=1000
M=10000
SumList = zeros(M)
stdlist = zeros(M)
randlist = zeros(N)
RandomStd = 0.0
NormalStd = 0.0
TempList = zeros(N)
function randgenerator(n)           # Generating n Random numbers between 1 and 10 
    randlist = rand(1:10 , n)
    return(randlist)
end

for i in 1:M
    templist = randgenerator(N)             # Calculating mean and standard deviation for fitting a normal distribution
    SumList[i] = sum(templist)
    stdlist[i] = std(templist)
end

NormalStd = sum(stdlist) / M*sqrt(N)

p1 = stephist(SumList , normalize=:pdf, label = L"count", c = :plum , fill = true)              #PLotting
p2 = plot!(Normal(sum(SumList)/M,NormalStd), label = L"fitted\ normal\ distribution", c = :darkorchid3)
plot!(xlabel = L"sum\ of\ randlists", ylabel = L"density",title = L"Sum\ of\ M=10^4\ randlists and N=10^3")
display(p1)
savefig(p1,"normalN1000-M10^4")




```
