```julia
using LinearAlgebra
using Plots
using Random
using Statistics
using LaTeXStrings
```

```julia
N = 100000
CountList = zeros(Int , 10)
linelist = ones(Int , 10)
xlist = collect(1:10)
randlist1 = rand(1:10 , N)
randlist2 = fill(11,N)
index = 0
temp = 1
for i in 1:N                  #Finding the values after every 4 and saving it in ana array
     if randlist1[i] == 4
        index = index + 1 
        randlist2[index] = randlist1[i+1]
     end
 end

for i in 1:index                          # Counting the nubers od times each value is repeated since we are not using histogram function for plotting
   CountList[randlist2[i]] = CountList[randlist2[i]] + 1
end

p1 = bar( xlist , CountList )       #Plotting
p2 = plot!(xlist, linelist*(index/10),xlabel = L"number" ,ylabel = L"count", title = L"uniform\ distribution\ histogram\ for\ N=100000")
display(p2)
savefig("images\\6.2.png")

```
