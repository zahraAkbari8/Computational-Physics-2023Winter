```python
import matplotlib.pyplot as plt
import numpy as np
import random as rnd
from scipy import stats
import statistics
from scipy.optimize import curve_fit
```

```python
t = 10000
L = 200
h = np.zeros(200, dtype=int)
w = np.zeros(t)
h_sum = 0
h2_sum = 0
h_mean2 = 0
h2_mean = 0
plt.figure(figsize=((200*2.9)/300, (200*2.9)/300), dpi=300)

colorlist= ['purple','#BF3EFF','#A2CD5A','#F08080']
for i in range(t):
    color_i = (i//2500) % 4                                     #dedicating different colors fro, colorlist for each 2500 squares
    x = rnd.randint(0,199)                                      #choosing a random number betweeen 0 and 199 (h imdex goes from 0 to 199)
    h[x] += 1                                                   #adding 1 to a random index of h
    plt.scatter(x,h[x]/4, s=[2.9], marker='s', color = colorlist[color_i],zorder=color_i)      #plotting by scattering small squares
    
    h_mean2 = (i/L)*(i/L)                                       #calculating standard deviation 
    for j in range(L):
        h2_sum = h2_sum+ (h[j])^2
    h2_mean = h2_sum/L
    w[i] = np.sqrt(h2_mean-h_mean2)
    

ax = plt.gca()
ax.set_aspect(aspect=10)
#plt.axis('scaled')
plt.xlim(0, 200)
plt.ylim(0, 20)
plt.savefig('scatter3',figsize=((200*2.9)/300, (200*2.9)/300), dpi=300)
plt.show()    

```

```python

t= np.linspace(1,10000,10000)
plt.figure(dpi=300)
plt.scatter(t,w,s=[1],color ='#A2CD5A')                     #plotting standard deviation(t) in different times.
ax = plt.gca()

ax.set_yscale('log')                                        #changing to logarithmic scale to see the linear property
ax.set_xscale('log')

def f(x, A, B):                                             # this is a straight line y=f(x)
    return A*x + B

popt, pcov = curve_fit(f, (t), w)         #fitting a line using Least square method from scipy.optimize library
print(popt[0])
p = np.poly1d([popt[0], popt[1]])
plt.plot(t, p(t),color='#68228B',linewidth=0.8, label=f'$y = {round(popt[0],6)} {round(popt[1],6):+}$')     #popt[0] is the slope and popt[1] is the intercept
plt.legend()
ax.set_xlabel(f't')
ax.set_ylabel(f'w(t)')

plt.savefig('line1', dpi=600)
plt.show
```
