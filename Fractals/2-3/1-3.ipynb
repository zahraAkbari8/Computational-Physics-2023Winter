```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
k = 1/2
```

```python
#The following lines are copied from problem1-1
#This function contracts the lines with ratio r.
def contract(coordinates,r):
    newcoordinates = []
    for i in range(len(coordinates)):
        p = coordinates[i] * r
        newcoordinates.append(p)
    return(newcoordinates)

#This function moves the line in 2D space. where x is the amount by which the line is moved horizontally.
def translation(coordinates,x,y):
    newcoordinates=[]
    for i in range(len(coordinates)):
        p = coordinates[i]  +  np.array([x,y])
        newcoordinates.append(p)
    return(newcoordinates)
```

```python
# contracts by 1/2.
def function1(coordinates):
    T1 = contract(coordinates,k)
    return(T1)

# contracts by 1/2, moves to the right by 1/2
def function2(coordinates):
    T2 = contract(coordinates,k)
    T2 = translation(T2,1/2,0)
    return(T2)

# contracts by 1/2, moves to the right by 1/4 and upward by sqrt(3)/4
def function3(coordinates):
    T3 = contract(coordinates,k)
    T3 = translation(T3,1/4,np.sqrt(3)/4)
    return(T3)

# combining functions
def Function(coordinates):
    T1 = function1(coordinates)
    T2 = function2(coordinates)
    T3 = function3(coordinates)
    T = T1+T3+T2
    return(T)
```

```python
#starting point with a triangle with following co-ordinates.
triangle0 = np.array([[1/2,np.sqrt(3)/2],[1,0],[0,0],[1/2,np.sqrt(3)/2]])
for i in range(10):
    triangle0 = Function(triangle0)
#we want to plot (colour) each triangle at a time so that unnecessary lines are not produces.
# so we are separting the points of each triangle (the last point is repeated so that we can draw all edges)
length = len(triangle0)
n = int(length/4)
q = length / n
plt.figure(dpi=300)
#plotting (filling) the triangles
for i in range(n):
    x = [triangle0[j][0] for j in range(4*i,4*i+4)]
    y = [triangle0[j][1] for j in range(4*i,4*i+4)]
    plt.fill(x, y)
plt.savefig('Serpinski 10 ', dpi=300, linewidth =0.01 )
```
