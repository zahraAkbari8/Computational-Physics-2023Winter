```python
import matplotlib.pyplot as plt
import numpy as np
import random as rnd
import seaborn as sns
c =-0.4+0.6*1j
```

```python
#defining function f= z^2+c
def function(coordinates):
    coordinates = coordinates*coordinates + c
    return(coordinates)
```

```python
colorvalue =0                                                       
arr = np.zeros((3000,3000))                             

for i in range(3000):                                        #dividing the 2d plane between -1.5 and 1.5 to 3000*3000 parts
    for j in range(3000):
        counter = 0
        a = ((i-1500)/1000) 
        b = ((j-1500)/1000) 
        r = a + b * 1j

        while (abs(r)<=2 and counter<200):                   #if the function has'nt diverged yet 
            r = function(r)
            colorvalue = counter/200                         #counting in a loop the steps needed for it to diverge. 
            arr[i][j] = colorvalue                           #setting a color based on the steps for the heatmap
            counter = counter+1                              

plt.figure(dpi=600)
axis_amounts = np.linspace(0,3000,7)                                #labellling the right amounts for axis labels 
new_axis_amount_x = [-1.5,-1,-0.5,0,0.5,1,1.5]
new_axis_amount_y = [1.5,1,0.5,0,-0.5,-1,-1.5]
plt.imshow( arr , cmap = 'viridis' , interpolation = 'nearest' )
plt.title( "Julia Set " )
ax = plt.gca()
ax.set_xticks(axis_amounts)
ax.set_xticklabels(new_axis_amount_x)
ax.set_yticks(axis_amounts)
ax.set_yticklabels(new_axis_amount_y)
plt.savefig('Julia8', dpi=600)
plt.show()
```
