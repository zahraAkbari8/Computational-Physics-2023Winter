```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
theta = -np.pi/4
theta2 = -3/4 * np.pi
k = np.sqrt(2)/2
```

```python
#The following lines are copied from problem1-1
#This function contracts the lines with ratio r.
def contract(coordinates,r):
    newcoordinates = []
    for i in range(len(coordinates)):
        p = coordinates[i] * r
        newcoordinates.append(p)
    return(newcoordinates)

#This function moves the line in 2D space. where x is the amount by which the line is moved horizontally.
def translation(coordinates,x,y):
    newcoordinates=[]
    for i in range(len(coordinates)):
        p = coordinates[i]  +  np.array([x,y])
        newcoordinates.append(p)
    return(newcoordinates)
#a function for rotating the line by (th) angle.
def rotation(coordinates,th):
    newcoordinates = []
    rotate = np.array([[np.cos(th),np.sin(-th)],[np.sin(th),np.cos(th)]])
    for i in range(len(coordinates)):
        p = np.matmul(rotate, coordinates[i]) 
        newcoordinates.append(p)
    return(newcoordinates)
```

```python
# contracts by coefficent k = np.sqrt(2)/2  rotates -45 degrees
def Line1(coordinates):
    L1 = contract(coordinates,k)
    L1 = rotation(L1 , theta)
    return(L1)

#  contracts by coefficent k = np.sqrt(2)/2, rotates -135 degrees and moves by 2 in horizontal axis.
def Line2(coordinates):
    L2 = contract(coordinates,k)
    L2 = rotation(L2 , theta2)
    L2 = translation(L2,2,0)
    L2.reverse()
    return(L2)

#combining the parts
def total_Line(coordinates):
    L1 = Line1(coordinates)
    L2 = Line2(coordinates)
    L = L1+L2
    return(L)


```

```python
points = np.array([[0,0],[1,-1],[2,0]])
for i in range(17):
    points = total_Line(points)
length = len(points)
#plotting the fractal

x1 = [item[0] for item in points[1:int(length/2)]]
y1 = [item[1] for item in points[1:int(length/2)]]
x2 = [item[0] for item in points[int(length/2):]]
y2 = [item[1] for item in points[int(length/2):]]
```

```python
plt.figure(dpi=300)
plt.plot(x1,y1,linewidth = 0.25, color='violet')
plt.plot(x2,y2,linewidth = 0.25, color = 'purple')
plt.savefig('heigh-way with 17 steps', dpi=300, linewidth =0.01 )
```
